```r
l1_pvclust <- pvclust(l1
            ,method.hclust = "ward.D"
            ,method.dist ="correlation"
            ,nboot = 1000
            ,iseed = 1)
plot(l1_pvclust)
pvrect(l1_pvclust)
```

### Кластеризація всіх параметрів

Мені стало цікаво чи зможе якийсь алгоритм кластеризації якось якісно визначити елементи з нашого датасету. Скомбінуємо всі датасети в один і застосуємо алгоритм кластеризації

```r
l1['class'] <- 1
l2['class'] <- 2
g1['class'] <- 3
```

```r
all <- rbind(l1,l2,g1)
```

```r
new<-all[,1:4]
```

```r
res <- kmeans(new,centers = 3,nstart = 20)
all['class_predict_kmeans'] <- res$cluster
```

```r
all1 <- dist(new,method = "euclidian")
plot(hclust(all1,method = "ward.D"))
```

```r
clusters <- cutree(hclust(all1,method = "ward.D"), 3)
table(clusters)
```

```r
new_list <- list()
for( i in 1:length(clusters)){
    if (clusters[[i]]==1){new_list[[i]]<-2 }
    if (clusters[[i]]==2){new_list[[i]]<-3 }
    if (clusters[[i]]==3){new_list[[i]]<-1 }
}
ordered_vector <- unlist(new_list)[order(unlist(new_list))]
```

```r
#all['hier'] <- unname()
table(all$class,ordered_vector)
```

```r
table(all$class,all$class_predict_kmeans)
```

Висновок:
Судячи з аналізу плеяд у нас глайдер схожий на лінарс2. 
А з аналізу  кластеризації обєднаних даних, можна сказати, що глайдер схожий на лінарс 2 більше ніж на лінарс 1
